## Filtering Data

```python
# Single condition
df[df["sex"] == 2]
```

```python
#Multiple conditions
df[(df["sex"] == 2) & (df["age"] >= 50)]
```

```python
#Excluding missing values
df[df["age"].notna()]
```

```python
#Filtering by numerical ranges
df[df["age"].between(55, 64)]
```

```python
#Filtering by .isin command
df[df["ethnic"].isin([1, 2, 3])]
```

```python
#Creating derived variables
df["age_55_plus"] = df["age"] >= 55
```
